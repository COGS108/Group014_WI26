## Abstract

Sleep deprivation is common among university students, yet its measurable impact on
academic performance remains underexplored in large-scale empirical work. This project
investigates whether students who sleep 7 or more hours per night achieve statistically
significantly higher cumulative GPAs than those who sleep fewer than 7 hours, using
Objective: Fitbit-measured sleep data from Carnegie Mellon University's CMU Sleep dataset
(n = 634 observations).

We conducted a complete data science workflow, including step-by-step data wrangling,
exploratory data analysis, and hypothesis testing using two complementary statistical
tests. Our primary analysis used a Welch's two-sample t-test, and we additionally ran
A Mann-Whitney U test to verify robustness given the left-skewed GPA distribution. We
also explored potential confounding variables, including course load, bedtime regularity,
and daytime napping behavior.

Our Welch's t-test returned t = 2.10, p = 0.036, allowing us to reject the null
hypothesis that there is a statistically significant difference in mean CGPA between the
two groups (3.517 vs. 3.441). The Mann-Whitney U test corroborated this result
(U = 48930, p = 0.025). However, Cohen's d ≈ 0.17 indicates a small practical effect
size, suggesting sleep is a genuine but modest predictor of academic performance.
Course load was ruled out as a confound (p = 0.867), and daytime napping did not
appear to compensate for insufficient nighttime sleep. These findings support the
7-hour threshold as a meaningful boundary for academic performance while acknowledging
that sleep is one of many contributing factors.

## Research Question

Is there a statistically significant difference in the average cumulative GPA (CGPA)
of university students who sleep 7 or more hours per night compared to those who sleep
fewer than 7 hours per night, and how much of the observed difference can be explained
by sleep duration alone after controlling for potential confounds such as course load
and sleep regularity?


## Background and Prior Work

The relationship between sleep hygiene and academic success is a cornerstone of student
health research. University students often experience a "sleep debt" due to high cognitive
demands and the biological tendency for delayed sleep phases in young adults. Research
indicates that sleep is not merely a passive state of rest but an active period for memory
consolidation, where the hippocampus and neocortex interact to solidify information
acquired during the day.1 When students drop below the recommended threshold of rest,
they risk impairing executive functions, including attention span and decision-making
capabilities, which are vital for maintaining a high CGPA.

The American Academy of Sleep Medicine emphasizes that for college students, getting enough
sleep is vital for academic success, noting that sleep-deprived students are more likely to
experience lower grades and a decrease in overall performance.2 While many students
prioritize "all-nighter" study sessions, data consistently shows that consistent,
long-duration sleep provides a much stronger foundation for long-term retention and
academic endurance.

Furthermore, large-scale longitudinal studies have demonstrated that sleep quality,
duration, and consistency are all significantly related to academic performance in college.
Specifically, students who maintain a regular sleep schedule and average over seven hours
of rest tend to have significantly higher GPAs than those with irregular or short-duration
sleep patterns.3 This suggests that the quality and quantity of rest may be a more powerful
predictor of success than the sheer volume of study hours completed.

By utilizing Exploratory Data Analysis on student lifestyle data, we aim to verify if a
7-hour threshold results in a statistically significant difference in CGPA. Organizations
like the Sleep Foundation have historically tracked how sleep affects school reports,
providing a baseline that suggests a clear drop-off in academic metrics when students fall
into the "insufficient sleep" category.4 Our project builds on this by specifically
isolating the 7-hour variable to provide an unambiguous answer to how rest influences
academic standing.

## Hypothesis


We predict that there will be a large, statistically significant, and positive difference
in the average CGPA of university students who report sleeping 7+ hours per night compared
to the average CGPA of university students who report sleeping less than 7 hours per night.
We believe that students who have more sleep will likely be more alert in the classes they
take, and as a result will receive higher scores on homework assignments and exams, on
average. This means we would expect their CGPAs to be greater compared to students who
sleep less.

**Null hypothesis (H₀):** There is no significant difference in mean CGPA between students
who sleep 7+ hours and those who sleep fewer than 7 hours.

**Alternative hypothesis (H₁):** Students who sleep 7+ hours have a significantly higher
mean CGPA than those who sleep fewer than 7 hours.

We use a significance threshold of α = 0.05.

## Data

### Data Overview

We use one primary dataset: the **CMU Sleep dataset**
(https://www.kaggle.com/datasets/mexwell/cmu-sleep). This dataset was collected at
Carnegie Mellon University where first-year students wore Fitbit devices for one month
during an academic term. Sleep metrics were paired with institutional GPA records.
Each row represents one student during one academic term (634 total observations,
15 variables).

**Key variables used in our analysis:**
- `TotalSleepTime`: Average nightly sleep in minutes — our primary independent variable
- `cum_gpa`: Cumulative GPA (0.0–4.0) — our primary dependent variable
- `term_gpa`: End-of-term GPA (0.0–4.0) — supplementary outcome variable
- `bedtime_mssd`: Day-to-day variability in bedtime (sq. hours) — sleep regularity measure
- `daytime_sleep`: Average daytime/nap sleep in minutes
- `Zterm_units_ZofZ`: Standardized course load relative to peers — tested as a potential
  confound

The dataset does not directly have a variable indicating whether an individual's average
sleep exceeded 7 hours, but we created this during the cleaning phase.

**Limitations:** The individuals in the dataset opted into the Fitbit study, which
introduces self-selection bias — participants may be more health-conscious or academically
motivated than average students. Furthermore, there may be some measurement error with the
Fitbits (e.g., if a student's Fitbit died mid-sleep or if they forgot to wear it), and we
are missing `term_units` values for 147 students, though this does not affect our key
variables.

Test to compare the distributions of the two groups in a way that does not assume
normality.

### EDA Section 2 — Hypothesis Test: Do 7+ Hour Sleepers Have Higher CGPA?

Our research question asks whether sleeping 7+ hours produces a *statistically significant*
difference in CGPA — not just an apparent one that could arise from random sampling
variation. We use both a Welch's two-sample t-test, which quantifies the probability that
the observed mean difference is due to chance alone, and a Mann-Whitney U test, which is
nonparametric and does not assume the distribution of the variable.

We chose Welch's t-test over Student's t-test because it does not assume equal variances
between groups, which is more plausible given our unequal group sizes (206 vs. 428
students).

```python
# Separate into two groups
group_high = cmu_clean[cmu_clean['sleep_group'] == '7+ Hours']['cum_gpa']
group_low  = cmu_clean[cmu_clean['sleep_group'] == '< 7 Hours']['cum_gpa']

# Summary statistics per group
print("--- Group Summary Statistics ---")
summary = cmu_clean.groupby('sleep_group')['cum_gpa'].agg(['count', 'mean', 'std', 'median'])
summary.columns = ['N', 'Mean GPA', 'Std Dev', 'Median GPA']
print(summary.round(4))

# Welch's t-test
t_stat, p_val = stats.ttest_ind(group_high, group_low, equal_var=False)

# Cohen's d effect size
pooled_std = np.sqrt((group_high.std()**2 + group_low.std()**2) / 2)
cohen_d    = (group_high.mean() - group_low.mean()) / pooled_std

print(f"\n--- Welch's T-Test Results ---")
print(f"T-statistic : {t_stat:.4f}")
print(f"P-value     : {p_val:.4e}")
print(f"Cohen's d   : {cohen_d:.4f}  (small < 0.2, medium ~ 0.5, large > 0.8)")
print(f"Mean diff   : {group_high.mean() - group_low.mean():.4f} GPA points (7+ hrs minus <7 hrs)")
print("\nConclusion  :",
      "Statistically significant (p < 0.05)" if p_val < 0.05 else "Not significant")
```

**Interpretation — Welch's T-Test:**

We obtained t = 2.10, p = 0.036, below our significance threshold of α = 0.05. We
therefore **reject the null hypothesis** — there is statistically significant evidence of
a difference in mean CGPA between students sleeping 7+ hours and those sleeping fewer
than 7 hours.

Students in the 7+ Hours group average 3.517 CGPA vs. 3.441 for the < 7 Hours group,
a difference of ~0.076 GPA points. However, Cohen's d ≈ 0.17 indicates a **small
practical effect size**. While the difference is real and unlikely due to chance, sleep
alone is a modest predictor of GPA; other factors such as study habits, course difficulty,
or prior academic ability likely account for much more of the variance.

#### The Mann-Whitney U Test

Because our density plot indicates that cumulative GPA is left-skewed, we also perform a
Mann-Whitney U test, which is nonparametric and does not assume the distribution of the
variable. The assumptions for the Mann-Whitney U Test are that the observations in each
group are independent of each other, and that the responses are ordinal. These assumptions
are met once we create the rank-transformed variable below.

Our null hypothesis under the Mann-Whitney U Test is that the distribution of cumulative
GPAs is equal between our two groups. The alternative hypothesis is that the distributions
are not equal.

```python
# Creating the rank-transformed cumulative GPA variable for the U test
cmu_clean['cum_gpa_rank'] = stats.rankdata(cmu_clean['cum_gpa'], method='ordinal')
group_high_u = cmu_clean[cmu_clean['sleep_group'] == '7+ Hours']['cum_gpa_rank']
group_low_u  = cmu_clean[cmu_clean['sleep_group'] == '< 7 Hours']['cum_gpa_rank']

u_stat, u_test_p_val = stats.mannwhitneyu(group_high_u, group_low_u)

print("\n--- Mann-Whitney U-Test Results ---")
print(f"U-statistic : {u_stat:.4f}")
print(f"P-value     : {u_test_p_val:.4e}")
print("\nConclusion  :",
      "Statistically significant (p < 0.05)" if u_test_p_val < 0.05 else "Not significant")
```

**Interpretation — Mann-Whitney U Test:**

We obtained U = 48930, p = 0.025. We reject the null hypothesis — there is statistically
significant evidence that the distribution of cumulative GPAs between our two groups are
not equal. This corroborates the results of the Welch's t-test and confirms that the
finding is robust to the non-normality of the GPA distribution.

A key confounding question: do students who sleep less do so because they carry heavier

We test this by comparing standardized course loads (`Zterm_units_ZofZ`) between groups.

```python
# Ensure course load is numeric
cmu_clean['Zterm_units_ZofZ'] = pd.to_numeric(cmu_clean['Zterm_units_ZofZ'], errors='coerce')

palette = {'7+ Hours': '#2ecc71', '< 7 Hours': '#e74c3c'}
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 4A: Course load boxplot
sns.boxplot(
    x='sleep_group', y='Zterm_units_ZofZ', data=cmu_clean,
    hue='sleep_group', palette=palette,
    order=['7+ Hours', '< 7 Hours'], ax=axes[0]
)
axes[0].axhline(0, color='gray', linestyle='--', linewidth=1, label='Population avg (Z=0)')
axes[0].set_title('Course Load by Sleep Group', fontsize=13)
axes[0].set_xlabel('Sleep Group', fontsize=11)
axes[0].set_ylabel('Standardized Course Load (Z-score)', fontsize=11)
axes[0].legend()

# Plot 4B: Course load vs CGPA scatter
for group, color in palette.items():
    subset = cmu_clean[cmu_clean['sleep_group'] == group]
    axes[1].scatter(subset['Zterm_units_ZofZ'], subset['cum_gpa'],
                   alpha=0.3, label=group, color=color, s=30)
axes[1].set_title('Course Load vs. CGPA by Sleep Group', fontsize=13)
axes[1].set_xlabel('Standardized Course Load (Z-score)', fontsize=11)
axes[1].set_ylabel('Cumulative GPA', fontsize=11)
axes[1].legend(title='Sleep Group')

plt.suptitle('Figure 4 — Course Load as a Potential Confound', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# Welch t-test: course load difference between groups
cl_high = cmu_clean.loc[cmu_clean['sleep_group'] == '7+ Hours', 'Zterm_units_ZofZ'].dropna()
cl_low  = cmu_clean.loc[cmu_clean['sleep_group'] == '< 7 Hours', 'Zterm_units_ZofZ'].dropna()
t_cl, p_cl = stats.ttest_ind(cl_high, cl_low, equal_var=False)

print(f"N (7+ hrs): {len(cl_high)}")
print(f"N (<7 hrs): {len(cl_low)}")
print(f"Mean course load — 7+ hrs: {cl_high.mean():.3f}")
print(f"Mean course load — <7 hrs: {cl_low.mean():.3f}")
print(f"Welch's t-test: t = {t_cl:.3f}, p = {p_cl:.4f}")
if p_cl < 0.05:
    print("Conclusion: Groups differ in course load.")
else:
    print("Conclusion: No significant course load difference — confound ruled out.")
```

### Linear Regression: Predicting CGPA from Sleep and Related Variables

While our hypothesis tests confirmed a statistically significant difference in mean CGPA
between sleep groups, they do not tell us how much of the variation in CGPA can be
explained by sleep duration, or how the relationship holds when we account for other
variables simultaneously. To address this, we fit a series of Ordinary Least Squares (OLS)
linear regression models predicting cumulative GPA from sleep-related features.

We fit three models in increasing order of complexity:
1. **Model 1 (baseline):** `cum_gpa ~ TotalSleepTime` — sleep duration alone
2. **Model 2 (+ regularity):** `cum_gpa ~ TotalSleepTime + bedtime_mssd` — adding sleep
   schedule regularity
3. **Model 3 (full):** `cum_gpa ~ TotalSleepTime + bedtime_mssd + daytime_sleep +
   Zterm_units_ZofZ` — all available predictors

This allows us to see whether sleep duration retains its predictive value after controlling
for sleep regularity, daytime napping, and course load, and to quantify how much variance
in CGPA each model explains.

```python
## YOUR CODE HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION
import statsmodels.formula.api as smf

# Ensure all columns are numeric
cmu_clean['Zterm_units_ZofZ'] = pd.to_numeric(cmu_clean['Zterm_units_ZofZ'], errors='coerce')

# Model 1: Sleep duration only
lm1 = smf.ols("cum_gpa ~ TotalSleepTime", data=cmu_clean).fit()

# Model 2: Sleep duration + bedtime regularity
lm2 = smf.ols("cum_gpa ~ TotalSleepTime + bedtime_mssd", data=cmu_clean).fit()

# Model 3: All predictors
lm3 = smf.ols(
    "cum_gpa ~ TotalSleepTime + bedtime_mssd + daytime_sleep + Zterm_units_ZofZ",
    data=cmu_clean
).fit()

# Summary comparison table
print("=" * 60)
print("Model Comparison Summary")
print("=" * 60)
print(f"{'Model':<45} {'R²':>6}  {'Adj R²':>8}  {'F p-val':>10}")
print("-" * 60)
print(f"{'Model 1: TotalSleepTime only':<45} {lm1.rsquared:.4f}  {lm1.rsquared_adj:.6f}  {lm1.f_pvalue:.4e}")
print(f"{'Model 2: + bedtime_mssd':<45} {lm2.rsquared:.4f}  {lm2.rsquared_adj:.6f}  {lm2.f_pvalue:.4e}")
print(f"{'Model 3: + daytime_sleep + Zterm':<45} {lm3.rsquared:.4f}  {lm3.rsquared_adj:.6f}  {lm3.f_pvalue:.4e}")
print("\n\n--- Model 1 Summary ---")
print(lm1.summary())
```

**Interpretation — Model 1 (TotalSleepTime only):**

The baseline regression confirms the relationship seen in our t-test. The coefficient on
`TotalSleepTime` is small but positive and statistically significant (p < 0.05), indicating
that each additional minute of nightly sleep is associated with a small increase in
cumulative GPA. However, the R² is very low (~0.02–0.04), meaning that sleep duration
alone explains only 2–4% of the variance in CGPA. This is consistent with Cohen's d ≈ 0.17
from our t-test: a real but modest relationship.

```python
print("--- Model 3 (Full Model) Summary ---")
print(lm3.summary())
```

**Interpretation — Model 3 (Full Model):**

Adding bedtime regularity, daytime sleep, and course load increases R² only modestly,
confirming that none of these variables are strong standalone predictors of CGPA either.
Notably, `Zterm_units_ZofZ` (course load) remains non-significant in the full model,
consistent with our confound analysis in EDA Section 5. `TotalSleepTime` retains a
positive and significant coefficient even after controlling for all other variables,
which strengthens our confidence that the sleep–GPA relationship is not explained away
by the variables we measured.

The low overall R² across all three models is an important finding in itself: sleep-related
variables explain only a small fraction of GPA variance. This tells us that while sleep is
a genuine predictor of academic performance, it operates within a much larger web of
factors that are not captured in this dataset.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 5A: Actual vs Predicted from Model 3
predicted = lm3.fittedvalues
residuals = lm3.resid

axes[0].scatter(predicted, cmu_clean['cum_gpa'].loc[predicted.index],
               alpha=0.3, color='steelblue', s=25)
axes[0].plot([predicted.min(), predicted.max()],
             [predicted.min(), predicted.max()],
             color='red', linestyle='--', linewidth=2, label='Perfect fit')
axes[0].set_title('Model 3: Predicted vs. Actual CGPA', fontsize=13)
axes[0].set_xlabel('Predicted CGPA', fontsize=11)
axes[0].set_ylabel('Actual CGPA', fontsize=11)
axes[0].legend()

# Plot 5B: Residuals vs Fitted
axes[1].scatter(predicted, residuals, alpha=0.3, color='mediumpurple', s=25)
axes[1].axhline(0, color='red', linestyle='--', linewidth=2)
axes[1].set_title('Model 3: Residuals vs. Fitted Values', fontsize=13)
axes[1].set_xlabel('Fitted Values', fontsize=11)
axes[1].set_ylabel('Residuals', fontsize=11)

plt.suptitle('Figure 5 — Linear Regression Diagnostics', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

### Bootstrap Confidence Intervals: Estimating the Reliability of the Mean GPA Difference

Our Welch's t-test demonstrated a statistically significant mean GPA difference between
sleep groups (p = 0.036), but p-values alone do not tell us about the precision of our
estimate. A confidence interval answers a different and complementary question: what is
the plausible range of the true mean GPA difference in the population?

We use **bootstrap resampling** (10,000 iterations) to construct an empirical 95%
confidence interval for the difference in mean CGPA between the 7+ Hours and < 7 Hours
groups. Bootstrapping is particularly appropriate here because it makes no assumptions
about the shape of the underlying GPA distribution — important given that we observed
left-skewness in our EDA.

We also run a **permutation test** as an additional check: if we randomly shuffle the
sleep group labels 10,000 times and recompute the mean difference each time, how often
do we observe a difference as large as what we actually found? This gives us a
non-parametric p-value that does not rely on the t-distribution assumption.

```python
## YOUR CODE HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION
np.random.seed(42)
n_iterations = 10000

group_high_vals = cmu_clean[cmu_clean['sleep_group'] == '7+ Hours']['cum_gpa'].values
group_low_vals  = cmu_clean[cmu_clean['sleep_group'] == '< 7 Hours']['cum_gpa'].values

observed_diff = group_high_vals.mean() - group_low_vals.mean()

# Bootstrap confidence interval
bootstrap_diffs = []
for _ in range(n_iterations):
    sample_high = np.random.choice(group_high_vals, size=len(group_high_vals), replace=True)
    sample_low  = np.random.choice(group_low_vals,  size=len(group_low_vals),  replace=True)
    bootstrap_diffs.append(sample_high.mean() - sample_low.mean())

bootstrap_diffs = np.array(bootstrap_diffs)
ci_lower = np.percentile(bootstrap_diffs, 2.5)
ci_upper = np.percentile(bootstrap_diffs, 97.5)

print(f"Observed mean difference (7+ hrs minus <7 hrs): {observed_diff:.4f} GPA points")
print(f"Bootstrap 95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"CI excludes 0: {ci_lower > 0}")
```

```python
# Permutation test
all_vals = np.concatenate([group_high_vals, group_low_vals])
n_high   = len(group_high_vals)

permutation_diffs = []
for _ in range(n_iterations):
    shuffled = np.random.permutation(all_vals)
    perm_high = shuffled[:n_high]
    perm_low  = shuffled[n_high:]
    permutation_diffs.append(perm_high.mean() - perm_low.mean())

permutation_diffs = np.array(permutation_diffs)
perm_p_val = np.mean(np.abs(permutation_diffs) >= np.abs(observed_diff))

print(f"\nPermutation test p-value (two-tailed): {perm_p_val:.4f}")
print("Conclusion:",
      "Significant (p < 0.05)" if perm_p_val < 0.05 else "Not significant")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 6A: Bootstrap distribution
axes[0].hist(bootstrap_diffs, bins=60, color='steelblue',
             edgecolor='white', alpha=0.8)
axes[0].axvline(observed_diff, color='red', linewidth=2,
               label=f'Observed diff = {observed_diff:.4f}')
axes[0].axvline(ci_lower, color='orange', linestyle='--', linewidth=2,
               label=f'95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]')
axes[0].axvline(ci_upper, color='orange', linestyle='--', linewidth=2)
axes[0].axvline(0, color='black', linestyle='-', linewidth=1.5,
               label='No difference (0)')
axes[0].set_title('Bootstrap Distribution of Mean GPA Difference', fontsize=13)
axes[0].set_xlabel('Mean GPA Difference (7+ hrs minus <7 hrs)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].legend(fontsize=9)

# Plot 6B: Permutation distribution
axes[1].hist(permutation_diffs, bins=60, color='mediumseagreen',
             edgecolor='white', alpha=0.8)
axes[1].axvline(observed_diff, color='red', linewidth=2,
               label=f'Observed diff = {observed_diff:.4f}')
axes[1].axvline(-observed_diff, color='red', linewidth=2, linestyle='--')
axes[1].set_title('Permutation Test: Null Distribution', fontsize=13)
axes[1].set_xlabel('Mean GPA Difference (permuted)', fontsize=11)
axes[1].set_ylabel('Frequency', fontsize=11)
axes[1].legend(fontsize=9)
axes[1].annotate(f'Permutation p = {perm_p_val:.4f}',
                xy=(0.05, 0.92), xycoords='axes fraction', fontsize=11, color='navy')

plt.suptitle('Figure 6 — Bootstrap CI and Permutation Test', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

**Interpretation — Figure 6:**

**Bootstrap Distribution (left):** The bootstrap distribution of the mean GPA difference
is centered at the observed value (~0.076 GPA points) and is roughly normally distributed,
which is expected by the Central Limit Theorem. The 95% confidence interval is
approximately [0.006, 0.148] GPA points. Critically, this interval **does not include 0**,
meaning we can be 95% confident that the true population-level mean GPA difference between
sleep groups is positive (favoring the 7+ Hours group). This is a meaningful complement to
the p-value from our Welch's t-test: not only is the difference statistically significant,
but the bootstrap CI tells us the plausible magnitude of the effect.

**Permutation Test (right):** The null distribution of mean differences under random group
assignment is centered at 0, as expected. Our observed difference (red solid line) falls in
the tail of this distribution, and the two-tailed permutation p-value corroborates the
parametric result from the Welch's t-test. This non-parametric confirmation is important
because it does not rely on the assumption of normally distributed GPAs — and given the
left-skewed GPA distribution we observed in EDA Section 1, having this additional
confirmation strengthens our overall conclusion.

Together, the t-test, Mann-Whitney U test, bootstrap CI, and permutation test all point
to the same conclusion: students sleeping 7+ hours per night have modestly but reliably
higher CGPAs than those sleeping fewer than 7 hours.
