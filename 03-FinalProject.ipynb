## Abstract

Sleep deprivation is a common issue among university students, yet its measurable
impact on academic performance remains underexplored in large-scale empirical work.
This project investigates whether students who sleep 7 or more hours per night achieve
statistically significantly higher cumulative GPAs than those who sleep fewer than 7 hours
hours, using objective Fitbit-measured sleep data from Carnegie Mellon University's
CMU Sleep dataset (n = 634).

We conducted a complete data science workflow, including data wrangling, exploratory
data analysis, hypothesis testing, and predictive modeling. Our primary analysis used
a Welch's two-sample t-test comparing the cumulative GPAs of the two sleep groups.
We also explored potential confounding variables, including course load, sleep
regularity (bedtime variability), and daytime napping behavior. A linear regression
model was trained to predict CGPA from sleep-related features.

Our results show a statistically significant difference in mean CGPA between the two
groups (t = 2.10, p = 0.036), with students sleeping 7+ hours averaging a CGPA of
3.517 compared to 3.441 for those sleeping fewer than 7 hours. However, Cohen's d ≈
0.17 indicates a small practical effect size, suggesting that while sleep is a genuine
predictor of academic performance, it is one of many contributing factors. Course load
was ruled out as a confound, and bedtime regularity showed a weak additional
association with GPA. These findings support the broader public health consensus that
adequate sleep benefits cognitive performance, while also highlighting the complexity
of predicting academic outcomes from any single behavioral variable.

## Research Question

Is there a statistically significant difference in the average cumulative GPA (CGPA)
of university students who sleep 7 or more hours per night compared to those who sleep
fewer than 7 hours per night, and how much of the observed difference can be explained
by sleep duration alone after controlling for potential confounds such as course load
and sleep regularity?


## Background and Prior Work

The relationship between sleep and academic performance is a well-studied area of
student health research. University students are particularly vulnerable to sleep
deprivation due to high cognitive demands, social pressures, and the biological
tendency for delayed sleep phases in young adults — a phenomenon known as "social
jet lag." Research indicates that sleep is not merely passive rest but an active
period of memory consolidation, during which the hippocampus and neocortex interact
to solidify information acquired during the day (Okano et al., 2019).

A landmark longitudinal study by Okano et al. (2019) examined sleep patterns in
college students using objective wearable data and found that sleep quality, duration,
and consistency were all significantly associated with academic performance. Students
who maintained consistent sleep schedules averaging over 7 hours per night had
measurably higher GPAs than those with irregular or insufficient sleep. Critically,
this study used the same CMU dataset we analyze here, providing direct methodological
grounding for our approach.

The American Academy of Sleep Medicine (AASM, 2017) recommends that adults aged
18–25 obtain 7–9 hours of sleep per night for optimal cognitive functioning.
The AASM notes that sleep-deprived college students are more likely to experience
decreased attention, impaired decision-making, and lower academic performance.
These effects are particularly pronounced during consolidation of procedurally and
declaratively demanding academic tasks such as exam preparation and essay writing.

Maheshwari and Shaukat (2019) further demonstrated in a study of medical students
that poor sleep quality was associated with significantly lower academic scores,
with effect sizes ranging from small to moderate depending on the metric used.
Their findings reinforce the hypothesis that the 7-hour threshold represents a
meaningful boundary, below which cognitive performance begins to measurably degrade.

Despite this body of evidence, most existing studies rely on self-reported sleep
data, which is subject to recall bias. Our project improves on this by using
Fitbit-measured objective sleep data, providing a more reliable measure of actual
sleep duration. We also extend prior work by explicitly testing for potential
confounding variables — including course load and sleep schedule regularity —
to isolate the independent contribution of sleep duration to CGPA.

## Hypothesis


## Hypothesis

We predict that university students who sleep 7 or more hours per night will have
a statistically significantly higher cumulative GPA than students who sleep fewer
than 7 hours per night. Specifically, we expect students with more sleep to be
more alert, attentive, and cognitively prepared during class, resulting in better
performance on assignments and exams and therefore higher CGPAs on average.

We operationalize the 7-hour threshold based on the American Academy of Sleep
Medicine's recommendation that adults obtain at least 7 hours per night for
optimal cognitive functioning.

**Null hypothesis (H₀):** There is no significant difference in mean CGPA between
students who sleep 7+ hours and those who sleep fewer than 7 hours.

**Alternative hypothesis (H₁):** Students who sleep 7+ hours have a significantly
higher mean CGPA than those who sleep fewer than 7 hours.

We use a significance threshold of α = 0.05.

## Data

### Data overview

We use one primary dataset for this project: the **CMU Sleep dataset**
(Kaggle: `mexwell/cmu-sleep`). This dataset was collected as part of a study at
Carnegie Mellon University in which first-year students wore Fitbit devices for
one month during an academic term. Objective sleep metrics were paired with
institutional academic records provided by the university. Each row represents
one student during one academic term, for a total of 634 observations.

This dataset is well-suited to our research question because it provides
**objective, device-measured sleep data** rather than self-reported estimates,
reducing recall bias. The key independent variable (`TotalSleepTime`) directly
measures nightly sleep duration in minutes, and the key dependent variable
(`cum_gpa`) measures cumulative academic performance — exactly the variables
needed to test our hypothesis.

**Key variables:**
- `TotalSleepTime`: Average minutes of nightly sleep (main sleep period) —
  our primary independent variable
- `cum_gpa`: Cumulative GPA (0.0–4.0) — our primary dependent variable
- `term_gpa`: End-of-term GPA (0.0–4.0) — supplementary outcome variable
- `bedtime_mssd`: Day-to-day variability in bedtime (square hours) —
  measure of sleep schedule regularity
- `daytime_sleep`: Average daytime/nap sleep in minutes
- `Zterm_units_ZofZ`: Standardized course load relative to peers —
  tested as a potential confound
- `demo_race`, `demo_gender`, `demo_firstgen`: Demographic controls

**Known limitations:** The sample is self-selected (students who opted into
the Fitbit study), which may skew toward health-conscious individuals and
limit generalizability. Fitbit accuracy is imperfect. Missing `term_units`
values (n = 147) do not affect our key variables, which are fully complete.

the data and informs our choice of statistical test.


# Plot 1A: Sleep duration distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].hist(df['TotalSleepTime'], bins=30, color='steelblue',
             edgecolor='white', alpha=0.85)
axes[0].axvline(420, color='red', linestyle='--', linewidth=2,
               label='7-hr threshold (420 min)')
axes[0].axvline(df['TotalSleepTime'].mean(), color='orange',
               linestyle='-', linewidth=2,
               label=f"Mean: {df['TotalSleepTime'].mean():.0f} min")
axes[0].set_title('Distribution of Nightly Sleep Duration', fontsize=13)
axes[0].set_xlabel('Total Sleep Time (minutes/night)', fontsize=11)
axes[0].set_ylabel('Number of Students', fontsize=11)
axes[0].legend()
```

# Plot 1B: CGPA distribution


```python
axes[1].hist(df['cum_gpa'], bins=30, color='mediumseagreen',
             edgecolor='white', alpha=0.85)
axes[1].axvline(df['cum_gpa'].mean(), color='orange', linestyle='-',
               linewidth=2, label=f"Mean: {df['cum_gpa'].mean():.2f}")
axes[1].axvline(df['cum_gpa'].median(), color='purple', linestyle='--',
               linewidth=2, label=f"Median: {df['cum_gpa'].median():.2f}")
axes[1].set_title('Distribution of Cumulative GPA', fontsize=13)
axes[1].set_xlabel('Cumulative GPA (0.0–4.0)', fontsize=11)
axes[1].set_ylabel('Number of Students', fontsize=11)
axes[1].legend()

plt.suptitle('Figure 1 — Distributions of Primary Variables', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('results/fig1_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

Only ~32% of students meet the recommended sleep duration. The CGPA distribution
is left-skewed, with most students clustered near 3.5–4.0 but a meaningful tail
toward lower values. The mean (3.46) falling below the median (3.55) confirms
the left skew, which we account for by reporting both means and medians in all
group comparisons.
